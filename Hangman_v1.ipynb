```python
from random import randrange
from google.colab import output     #this only works in google colab like this. Needs to be "from os import system" for other systems
import time

hangman={0:'''
        ____________
         |''',
        1:'''
        ____________
         |
         O''',
        2:'''
        ____________
         |
         O
        /''',
        3:'''
        ____________
         |
         O
        / \\''',
        4:'''
        ____________
         |
         O
        / \\
         |''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}


list_of_words = [
    "PYTHON", "BOOTCAMP", "PRIMER", "HANGMAN", "CODING",
    "TABLEAU", "SEQUEL", "LEARNING", "SCHOOL", "VIDEO", "DATA"
    ]
word = list_of_words[randrange(len(list_of_words))]
hidden_word = len(word) * "-"
mistake_list = []
mistake_count = 6

print('==============================================================')


while mistake_count > 0 and hidden_word.count("-") != 0:
    letter_count = 0
    print(hangman[6 - mistake_count])
    print(f"You have {mistake_count} wrong guesses left")
    print(f"Hidden word is {hidden_word}")
    print(f"Your incorrect guesses are {*mistake_list, }")
    print("Pick a letter and press ENTER:")
    picked_letter = input().capitalize()
    if picked_letter.isalpha() == False or len(picked_letter) > 1:
      print("Please pick only one letter")

    for letter in word:
        if letter == picked_letter:
            hidden_word = hidden_word[:letter_count] + picked_letter + hidden_word[letter_count+1:]

        letter_count += 1

    if word.count(picked_letter) > 0:
        print(f"Correct! The word contains the {picked_letter}.")
        print(f"Hidden word is {hidden_word}")
    else:
        mistake_count -= 1
        mistake_list.append(picked_letter)
        print(f"Oh no! The word does not contain the {picked_letter}.")
        print(f"You have {mistake_count} wrong guesses left")
        print(f"Hidden word is {hidden_word}")
        print(hangman[6 - mistake_count])
    print('==============================================================')
    time.sleep(3)
    output.clear()    #again, just for google colab. For other systems it would be os.system("cls") for windows, os.system("clear") for mac/linux


if mistake_count == 0:
  print(hangman[6 - mistake_count])
  print(f"OH NO! You lost. The word was {word}")
else:
  print(f"CONGRATULATIONS! You won. The word was {word}")
```
